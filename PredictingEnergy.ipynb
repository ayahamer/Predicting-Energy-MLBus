# KNN

Basic Model without tuning

```python
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn import metrics

y = ava['energy']
X = ava.iloc[0:4380, 0:292]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, shuffle=False)
#elected to divide up training and testing with 80% going to former and 20% going to later, with the years/chronological order
#being respected/maintained with the older 80% of data being trained upon and the most recent 20% used for testing from a range of 1994-2007
```

```python
#build the KNN pipeline using the scaler of RobustScaler followed by MinMaxScaler to establish a standardized range without concern for noise/anomalies influencing the scaling
knn = KNeighborsRegressor()
scaler1 = RobustScaler()
scaler2 = MinMaxScaler()
basicKNN = Pipeline([
    ('robust', scaler1),
    ('setRange', scaler2),
    ('knn', knn)
])

basicKNN.fit(X_train, y_train)
y_train_basicKNN = basicKNN.predict(X_train)
y_predictions_basicKNN = basicKNN.predict(X_test)
```

```python
print(metrics.mean_absolute_percentage_error(y_train,y_train_basicKNN))
```

```python
print(metrics.root_mean_squared_error(y_train,y_train_basicKNN))
print(metrics.root_mean_squared_error(y_test,y_predictions_basicKNN))
print(metrics.mean_absolute_percentage_error(y_test,y_predictions_basicKNN))
#compute root mean square error
#cardinality of energy (spectrum) is up to 300000000 with an average of 16,000,000
#thus, our accuracy is within the range of the squared magnitude at 10^13 indicating an acceptable accuracy rate
```

```python
basicKNN_final = basicKNN.fit(X, y)
```

KNN HPO with Cross-Validation

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV



k_folds = 4
#folds have to be based on years, with 2 years per fold
# Create a cross-validation object
kf = KFold(n_splits=k_folds, shuffle=False)

param_grid = {
    'n_neighbors': list(range(1, 2190, 100)),  # Range of k values to search
    #expand search space for Neighbours
    'p': [1, 2]  # 1 for Manhattan distance, 2 for Euclidean distance
}

grid_search_basic_KNN = GridSearchCV(knn, param_grid, cv=4)  # 4-fold cross-validation
grid_search_basic_KNN.fit(X_train, y_train)
best_p = grid_search_basic_KNN.best_params_['p']
best_k = grid_search_basic_KNN.best_params_['n_neighbors']
best_knn_model = grid_search_basic_KNN.best_estimator_

best_knn_model.fit(X_train, y_train)
best_knn_HPO_predictions = best_knn_model.predict(X_test)
```

```python
print("Training mse ", metrics.root_mean_squared_error(y_train, y_train_basicKNN))
print("Training mse ", metrics.mean_absolute_percentage_error(y_train, y_train_basicKNN))
```

```python
mse = metrics.root_mean_squared_error(y_test,best_knn_HPO_predictions)
mseperc = metrics.mean_absolute_percentage_error(y_test,best_knn_HPO_predictions)
print("Best K:", best_k)
print("Best p:", best_p)
print("Mean Squared Error on the test set:", mse)
print("Mean Absolute Error % on the test set:", mseperc)
```

KNN HPO with Cascade Validation

```python
from sklearn.model_selection import TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=4)
print(tscv)
```

```python
grid_search_cascade_KNN = GridSearchCV(estimator=knn, param_grid=param_grid, cv=tscv)
grid_search_cascade_KNN.fit(X_train, y_train)
best_cascade_p = grid_search_cascade_KNN.best_params_['p']
best_cascade_k = grid_search_cascade_KNN.best_params_['n_neighbors']
best_knn_cascade_model = grid_search_cascade_KNN.best_estimator_

best_knn_cascade_model.fit(X_train, y_train)
best_knn_cascade_HPO_predictions = best_knn_cascade_model.predict(X_test)
```

```python
mse = metrics.root_mean_squared_error(y_test, best_knn_cascade_HPO_predictions)
mseperc = metrics.mean_absolute_percentage_error(y_test,best_knn_cascade_HPO_predictions)
print("Best K:", best_k)
print("Best p:", best_p)
print("Mean Squared Error on the test set:", mse)
print("Mean Absolute Error % on the test set:", mseperc)
```
